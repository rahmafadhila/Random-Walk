```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 0
y = 0
step = 100
```

```python
def printCoor(x,y):
    text_file = open('Coordinate.txt', 'a')
    text_file.write('{} {} \n'.format(x, y))
```

```python
for i in range(0, step):
    n = np.random.randint(0, 100)
    if (n > 0 and n <= 13):
        #north
        y+= 1
        printCoor(x,y)
    elif (n > 13 and n <= 26):
        #northEast
        x+= 1
        y+= 1
        printCoor(x,y)
    elif (n > 26 and n <= 39):
        #east
        x+= 1
        printCoor(x,y)
    elif (n > 39  and n <= 52):
        #southEast
        x+= 1
        y-= 1
        printCoor(x,y)
    elif (n > 52 and n <= 64):
        #South
        y-= 1
        printCoor(x,y)
    elif (n > 64 and n <= 76):
        #southWest
        x-= 1
        y-= 1
        printCoor(x,y)
    elif (n > 76 and n <= 88):
        #west
        x-= 1
        printCoor(x,y)
    elif (n > 88 and n <= 100):
        x-= 1
        y+= 1
        printCoor(x,y)
```

```python
X = []
Y = []
for line in open('Coordinate.txt', 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])

plt.plot(X, Y)
plt.show()
```
